```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('../data/collision_data.csv')

# Group by selected features and count collisions

df_clean = df.dropna(subset=['C_HOUR', 'C_MNTH', 'C_WTHR', 'C_RCFG'])

# 3. Aggregate collisions by selected features
collision_counts = (
    df_clean
    .groupby(['C_HOUR', 'C_MNTH', 'C_WTHR', 'C_RCFG'])
    .size()
    .reset_index(name='collision_count')
)

nb_model = smf.glm(
    formula="collision_count ~ C_HOUR + C_MNTH + C_WTHR + C_RCFG",
    data=collision_counts,
    family=sm.families.NegativeBinomial()
).fit()

# 5. Display summary
print(nb_model.summary())



```
