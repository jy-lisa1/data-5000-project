This file produces simple 1D histograms/barcharts illustrating the value counts of our variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/cleaned_data.csv')

cols_to_convert = [col for col in df.columns if col != 'P_SEX']

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int32')
```

```python
plt.figure()
sns.countplot(data=df, x="C_YEAR")
plt.title("Collisions by Year")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
