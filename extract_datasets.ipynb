This file extracts 2 datasets from the original dataset, with only relevant information retained at the collision level and person level. Missing values are converted to NaN.

```python
import pandas as pd
```

```python
df = pd.read_csv('data/combined_data.csv')
```

```python
def convert_to_nan(column: str, unknown_strings: list[str]):
    """Convert unknown values from string (e.g. "UU") to NaN."""
    df[column] = df[column].replace(unknown_strings, pd.NA)
```

```python
# convert unknown values to NaN

convert_to_nan('C_MNTH', ['UU']) # month
convert_to_nan('C_WDAY', ['U']) # day of week
convert_to_nan('C_HOUR', ['UU']) # hour of day
convert_to_nan('C_VEHS', ['UU', '**']) # vehicles involved
convert_to_nan('C_CONF', ['UU', 'QQ']) # collision configuration
convert_to_nan('C_RCFG', ['UU', 'QQ']) # collision road configuration
convert_to_nan('C_WTHR', ['U', 'Q']) # weather
convert_to_nan('C_RSUR', ['U', 'Q']) # road surface
convert_to_nan('C_RALN', ['U', 'Q']) # road alignment
convert_to_nan('C_TRAF', ['UU', 'QQ']) # traffic control
convert_to_nan('V_ID', ['UU', '*']) # vehicle sequence number
convert_to_nan('V_TYPE', ['UU', 'QQ']) # vehicle type
convert_to_nan('V_YEAR', ['UUUU']) # vehicle year
convert_to_nan('P_ID', ['UU']) # person sequence number
convert_to_nan('P_SEX', ['U']) # sex
convert_to_nan('P_AGE', ['UU', 'NN']) # age
convert_to_nan('P_PSN', ['UU', 'QQ']) # person's position
convert_to_nan('P_ISEV', ['U', 'N']) # injury severity
convert_to_nan('P_SAFE', ['UU', 'NN', 'QQ']) # safety precautions taken
convert_to_nan('P_USER', ['U']) # road user class
```

```python
collisions_df = df[['C_YEAR', 'C_MNTH', 'C_WDAY', 'C_HOUR', 'C_SEV', 'C_VEHS', 'C_CONF',
       'C_RCFG', 'C_WTHR', 'C_RSUR', 'C_RALN', 'C_TRAF', 'C_CASE']]
collisions_df = collisions_df.drop_duplicates().reset_index(drop=True)
collisions_df.shape
```

```python
person_df = df[['V_ID', 'V_TYPE', 'V_YEAR', 'P_ID', 'P_SEX', 'P_AGE', 'P_PSN', 'P_ISEV', 
                'P_SAFE', 'P_USER', 'C_CASE']]
person_df = person_df.drop_duplicates().reset_index(drop=True)
person_df.shape
```

```python
collisions_df.to_csv('data/collision_data.csv', index=False)
person_df.to_csv('data/person_data.csv', index=False)
```
